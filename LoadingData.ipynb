```python
import pandas as pd
import numpy as np
```

```python
data  = pd.read_excel("/Users/macbookair/Desktop/Pandas Tutoriall soonnnü/Adidas US Sales Datasets.xlsx")
```

```python
data.head()
```

```python
data.shape
```

```python
# null değerlerden kurtulmak için dropna() fonksiyonunu kullanabiliriz.
data.dropna(axis = 0, thresh = 2, inplace = True) # Burada thresh argümanı silmeyeceğimiz satır veya sütun için  en az nannull sayısını ifade edr
data.dropna(axis = 1, thresh = 1, inplace = True) # axis 0 ise satır axis = 1 ise sütun demekir.

data.head()
```

```python
data.columns = data.loc[3] # verimizin sütun isimleri bir alt satırda veri olarak yazılmıştı onu sütun ismi olarak yazalım. 
data.drop(index = 3, inplace = True) # yazdıktan sonra o satırı silebiliriz.
data.head()
```

```python
data.set_index(pd.Series([s for s in range(data.shape[0])]), inplace = True) # index değerlerini kafam karışmaması için 0'dan başlayan değerlere atadım.
```
